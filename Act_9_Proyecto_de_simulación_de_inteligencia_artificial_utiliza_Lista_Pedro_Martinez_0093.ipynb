```python
# --- SIMULADOR DE PESCADOR CON IA ---

# Al inicio, la cubeta de pesca está vacía
cubeta = []

# Variables de estado del simulador
clima = "Soleado"
energia = 100

print("==================================================")
print(" ¡BIENVENIDO AL ASISTENTE DE PESCA CON IA v1.0! ")
print("==================================================")
print("Has activado a 'Pescabot', el robot pescador.")

# Ciclo principal del juego/simulador
continuar = "si"
while continuar == "si":
    print("\n------------------------------------------------")
    print("ESTADO ACTUAL:")
    print("- Peces en la cubeta:", cubeta)
    print("- Clima actual:", clima)
    print("- Energía de Pescabot:", energia, "%")
    print("------------------------------------------------")

    # Menú interactivo
    print("¿Qué acción deseas que realice la IA?")
    print("1. Analizar el entorno y Pescar")
    print("2. Cambiar el clima del simulador")
    print("3. Revisar y clasificar la pesca")
    print("4. Recargar energía de Pescabot")
    print("5. Salir del simulador")

    opcion = input("Selecciona una opción (1-5): ")

    # OPCIÓN 1: IA para Pescar basada en el clima y energía
    if opcion == "1":
        if energia < 20:
            print("\n[¡ALERTA!] Pescabot no tiene suficiente energía para lanzar la caña. ¡Recárgalo!")
        else:
            print("\n[IA Pensando] Analizando el clima...")
            energia = energia - 20 # Pescar gasta energía

            # La IA toma una decisión basada en la variable 'clima'
            if clima == "Soleado":
                print("-> El sol brilla. La IA detecta un Pez Dorado cerca de la superficie.")
                cubeta.append("Pez Dorado")
                print("¡Éxito! 'Pez Dorado' añadido a la cubeta.")
            elif clima == "Lluvioso":
                print("-> Está lloviendo. A los peces grandes les gusta la lluvia. ¡Un Pez Espada picó!")
                cubeta.append("Pez Espada")
                print("¡Éxito! 'Pez Espada' añadido a la cubeta.")
            elif clima == "Tormenta":
                print("-> ¡Hay tormenta! El mar está picado. La IA falló el tiro y solo pescó una Bota Vieja.")
                cubeta.append("Bota Vieja")
                print("'Bota Vieja' añadida a la cubeta... qué mala suerte.")
            else:
                print("-> Clima desconocido. La IA se confundió y pescó una Alga Marina.")
                cubeta.append("Alga Marina")

    # OPCIÓN 2: Cambiar el clima para ver cómo reacciona la IA
    elif opcion == "2":
        print("\nCambiar el clima del simulador:")
        print("- Escribe: Soleado")
        print("- Escribe: Lluvioso")
        print("- Escribe: Tormenta")
        nuevo_clima = input("Elige el nuevo clima: ")

        if nuevo_clima == "Soleado":
            clima = "Soleado"
            print("El clima ha cambiado a Soleado.")
        elif nuevo_clima == "Lluvioso":
            clima = "Lluvioso"
            print("El clima ha cambiado a Lluvioso.")
        elif nuevo_clima == "Tormenta":
            clima = "Tormenta"
            print("¡Cuidado! El clima ahora es de Tormenta.")
        else:
            clima = "Desconocido"
            print("Clima extraño detectado.")

    # OPCIÓN 3: Revisar lo que hay en la lista (Cubeta)
    elif opcion == "3":
        print("\n--- INSPECCIÓN DE LA CUBETA ---")
        if cubeta == []:
            print("La cubeta está vacía. ¡Ve a pescar primero!")
        else:
            print("Contenido de tu cubeta:", cubeta)
            print("Cantidad de objetos atrapados:", len(cubeta))

            # Una pequeña IA que evalúa si la pesca fue buena o mala
            if "Bota Vieja" in cubeta:
                print("Evaluación de la IA: Deberías limpiar tu cubeta, tienes basura.")
            elif "Pez Espada" in cubeta:
                print("Evaluación de la IA: ¡Excelente pesca! Tienes un espécimen raro.")
            else:
                print("Evaluación de la IA: Vas por buen camino, sigue pescando.")

    # OPCIÓN 4: Recargar energía
    elif opcion == "4":
        if energia >= 100:
            print("\nLa energía ya está al 100%. ¡Listo para trabajar!")
        else:
            energia = 100
            print("\nPescabot ha sido conectado a la corriente. ¡Energía restaurada al 100%!")

    # OPCIÓN 5: Salir del programa
    elif opcion == "5":
        continuar = "no"
        print("\nApagando a Pescabot. ¡Gracias por simular con nosotros!")

    # Validación por si eligen un número que no está en el menú
    else:
        print("\n[Error] Opción no válida. Intenta de nuevo con un número del 1 al 5.")

print("\n==================================================")
print("               PROGRAMA FINALIZADO                ")
print("---Programa realizado por Pedro Martinez NL = 38---")
print("==================================================")
```
