#### Lista $Pescador$



```python
# --- SIMULADOR DE PESCA INTERACTIVO ---

# 1. Configuración inicial
Pescador = []         # La lista donde se guardarán las capturas inicia vacía [1, 2]
energia = 50          # Recurso limitado para el pescador
jugando = True        # Control lógico del ciclo interactivo

print("--- BIENVENIDO AL MAR DE PYTHON ---")

# 2. Ciclo principal (Interactivo)
while jugando == True and energia > 0: # Se repite mientras haya energía [3]
    print(f"\n--- MENÚ DE PESCA ---")
    print(f"Energía disponible: {energia}")
    print(f"Número de capturas en tu lista 'Pescador': {len(Pescador)}") # [4]
    print("1. Intentar pescar (Uso de IA)")
    print("2. Ver mi lista de capturas (Pescador)")
    print("3. Terminar jornada y salir")

    opcion = int(input("Selecciona una acción (1-3): ")) # Entrada del usuario [5]

    # --- Lógica de la Simulación ---
    if opcion == 1: # [5]
        print("\nConfigurando el anzuelo...")
        cebo = int(input("Elige cebo (1: Gusano, 2: Señuelo): "))
        profundidad = int(input("¿A qué profundidad lanzas (1-20 metros)? "))

        # Proceso de decisión (IA básica con operadores lógicos y relacionales)
        if cebo == 1 and profundidad < 8: # Operador 'and' y '<' [6]
            print("IA del Pez: ¡Un gusano cerca de la superficie! ¡Es mío!")
            print("Resultado: ¡Capturaste una Trucha!")
            Pescador.append("Trucha") # Añade el elemento a la lista 'Pescador' [2]
        elif cebo == 2 and profundidad >= 12: # Operador '>=' [6]
            print("IA del Pez: Veo algo brillante en lo profundo...")
            print("Resultado: ¡Increíble! Capturaste un Pez Espada.")
            Pescador.append("Pez Espada") # [2]
        else:
            print("IA del Pez: No me interesa lo que hay en el agua.")
            print("Resultado: El pez ignoró tu anzuelo.")

        energia = energia - 10 # Operador aritmético de resta

    elif opcion == 2: # [5]
        print("\n--- CONTENIDO DE LA LISTA 'PESCADOR' ---")
        if len(Pescador) == 0: # Operador relacional '==' [6]
            print("Tu lista de capturas está vacía.")
        else:
            print(f"Capturas actuales: {Pescador}") # Muestra la lista completa [1]

    elif opcion == 3:
        print("Recogiendo las redes... ¡Hasta la próxima!")
        jugando = False # Rompe la condición del ciclo while

    else:
        print("IA Error: Opción no válida. Intenta de nuevo.")

# 3. Resumen final
if energia <= 0:
    print("\n--- ¡TE HAS AGOTADO! ---")
    print("Ya no tienes energía para seguir pescando.")

print(f"\nJornada terminada. Capturaste {len(Pescador)} peces en total.")
print("--- FIN DEL PROGRAMA ---")
```
